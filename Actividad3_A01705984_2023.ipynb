```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name = '2023')
```

```python
# Vemos con qué estamos trabajando
df.info()
```

```python
df.head()
```

```python
# Eliminamos las primeras 2 filas
df = df[2:]
df.head()
```

```python
# Hacemos que la primera fila sea la de títulos

df.rename(columns = df.iloc[0], inplace=True)
```

```python
df = df[1:]
df.head()
```

```python
df.isnull().sum()
# Se observa que existen columnas con muchos valores nulos
```

##Eliminamos valores nulos

```python
# Vamos a usar bfill
df1 =df.fillna(method="bfill")
df1
```

```python
df1.isnull().sum()
```

```python
# Como aún faltan valores nulos en la columna de póliza, se aplica ffill
```

```python
df2 =df1.fillna(method="ffill")
df2.isnull().sum()
```

```python
df2.shape
```

```python
# Cono se ve que toda la columna de "otros" son nulos, se elimina
df2 = df2.drop(columns = 'Otros')
```

```python
df2.isnull().sum()
```

```python
# Simplificamos
df = df2
```

```python
# Hacemos diagrama de caja del df
fig = plt.figure(figsize =(15, 8))
df.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show()
```

##Proceso para eliminar outliers

###Por desviación estándar

```python
# Por std dv
y=df
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
df_nuevo = df[(y<=Limite_Superior)&(y>=Limite_Inferior)]
df_nuevo
```

```python
df.columns
```

```python
# Eliminamos columnas con datos que no fueron arrojados en el punto antepasado
df_nuevo = df_nuevo.drop(['Fecha', 'Folio', 'UUID', 'RFC', 'Proveedor', 'TIPO GASTO',
       'Descripción', 'MP', 'FP', 'Tipo', 'Status', 'Poliza'], axis=1)
df_nuevo
```

```python
# Analizamos valores nulos
df_nuevo.isnull().sum()
```

```python
#Reemplazamos valores atípicos del dataframe con promedio
df_limpio=df_nuevo.copy()
df_limpio=df_limpio.fillna(round(df_nuevo.mean(),1))
df_limpio
```

```python
# Vemos que no haya nulos
df_limpio.isnull().sum()
```

```python
# Hacemos diagrama de caja del df
fig = plt.figure(figsize =(15, 8))
df_limpio.plot(kind='box', vert=False)
plt.title("Valores Atípicos del df_limpio")
plt.show()
```

```python
columnas = df[['Fecha', 'Folio', 'UUID', 'RFC', 'Proveedor', 'TIPO GASTO',
       'Descripción', 'MP', 'FP', 'Tipo', 'Status', 'Poliza']]
df_final_std = pd.concat([columnas, df_limpio], axis=1)
df_final_std
```

```python
# Mandamos a csv
df_final_std.to_csv('desviacion_estandar_2023.csv')
```

###Por rango intercuartílico

```python
# Eliminamos columnas con datos que no son útiles para el IQR
df2 = df.drop(['Fecha', 'Folio', 'UUID', 'RFC', 'Proveedor', 'TIPO GASTO',
       'Descripción', 'MP', 'FP', 'Tipo', 'Status', 'Poliza'], axis=1)
df2
```

```python
#Se encuentran cuartiles 0.25 y 0.75
y=df2

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Obtenemos datos limpios del Dataframe
df_iqr1 = df2[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
df_iqr1
```

```python
# Vemos valores nulos
df_iqr1.isnull().sum()
```

```python
#Reemplazamos valores atípicos con promedio
#Realizamos una copia del dataframe
df_iqr2=df_iqr1.copy()
df_iqr2=df_iqr2.fillna(round(df_iqr1.mean(),1))
df_iqr2
```

```python
# Hacemos diagrama de caja del df
fig = plt.figure(figsize =(15, 8))
df_iqr2.plot(kind='box', vert=False)
plt.title("Valores Atípicos del df_limpio")
plt.show()
```

```python
# Unimos las columnas que habíamos quitado
columnas = df[['Fecha', 'Folio', 'UUID', 'RFC', 'Proveedor', 'TIPO GASTO',
       'Descripción', 'MP', 'FP', 'Tipo', 'Status', 'Poliza']]
df_final = pd.concat([columnas, df_iqr2], axis=1)
df_final
```

```python
# Mandamos a csv
df_final.to_csv('iqr_2023.csv')
```
